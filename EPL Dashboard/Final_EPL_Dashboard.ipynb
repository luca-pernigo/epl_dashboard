```python
#this data is only for last season we want all 3 seasons

epl_teams=pd.read_csv("epl_teams.csv")
epl_only=epl_teams[epl_teams["tournament"]=="Premier League"]
epl_only['timestamp'] = pd.to_datetime(epl_only['date'])
epl_only22=epl_only[epl_only["timestamp"]>="2021-08-13"]
epl_only22=epl_only22[epl_only22["timestamp"]<="2022-05-22"]


teams=[i for i in epl_only["home"].unique()]


epl_only21=epl_only[epl_only["timestamp"]>="2020-09-12"]
epl_only21=epl_only21[epl_only21["timestamp"]<="2021-05-23"]


epl_only20=epl_only[epl_only["timestamp"]>="2019-08-9"]
epl_only20=epl_only20[epl_only20["timestamp"]<="2020-07-26"]


epl_only19=epl_only[epl_only["timestamp"]>="2018-08-10"]
epl_only19=epl_only19[epl_only19["timestamp"]<="2019-05-12"]


epl_only18=epl_only[epl_only["timestamp"]>="2018-08-11"]
epl_only18=epl_only18[epl_only18["timestamp"]<="2019-05-13"]


epl_only17=epl_only[epl_only["timestamp"]>="2018-08-13"]
epl_only17=epl_only17[epl_only17["timestamp"]<="2019-05-21"]
```

```python
teams22=[i for i in epl_only22["home"].unique()]
teams22

teams21=[i for i in epl_only21["home"].unique()]
teams21

teams20=[i for i in epl_only20["home"].unique()]
teams20

teams19=[i for i in epl_only19["home"].unique()]
teams19



epl_teams=(list(set(teams22) & set(teams21)& set(teams20)& set(teams19)))
epl_teams
```

```python

#select only teams that staid in the premier legue for last 3 years

epl_only22=epl_only22[(epl_only22['home'].isin(teams)) | (epl_only22['away'].isin(teams))]
epl_only21=epl_only21[(epl_only21['home'].isin(teams)) | (epl_only21['away'].isin(teams))]
epl_only20=epl_only20[(epl_only20['home'].isin(teams)) | (epl_only20['away'].isin(teams))]
epl_only19=epl_only19[(epl_only19['home'].isin(teams)) | (epl_only19['away'].isin(teams))]
```

```python
home_date22=epl_only22[["home", "timestamp","ht_result"]]

home_date21=epl_only21[["home", "timestamp","ht_result"]]

home_date20=epl_only20[["home", "timestamp","ht_result"]]

home_date19=epl_only19[["home", "timestamp","ht_result"]]
```

```python
def points_ht(row):
    if row['ht_result'] == "WON":
        val = 3
    elif row['ht_result'] == "DRAW":
        val = 1
    else:
        val = 0
    return val
```

```python
def points_at(row):
    if row['at_result'] == "WON":
        val = 3
    elif row['at_result'] == "DRAW":
        val = 1
    else:
        val = 0
    return val
```

```python
#don't know error but it works
home_date22['points'] = home_date22.apply(points_ht, axis=1)
home_date21['points'] = home_date21.apply(points_ht, axis=1)
home_date20['points'] = home_date20.apply(points_ht, axis=1)
home_date19['points'] = home_date19.apply(points_ht, axis=1)
```

```python
away_date22=epl_only22[["away", "timestamp","at_result"]]
away_date22['points'] = away_date22.apply(points_at, axis=1)
away_date22.columns=['home', 'timestamp', 'ht_result', 'points']
away_date22


away_date21=epl_only21[["away", "timestamp","at_result"]]
away_date21['points'] = away_date21.apply(points_at, axis=1)
away_date21.columns=['home', 'timestamp', 'ht_result', 'points']
away_date21


away_date20=epl_only20[["away", "timestamp","at_result"]]
away_date20['points'] = away_date20.apply(points_at, axis=1)
away_date20.columns=['home', 'timestamp', 'ht_result', 'points']

away_date19=epl_only19[["away", "timestamp","at_result"]]
away_date19['points'] = away_date19.apply(points_at, axis=1)
away_date19.columns=['home', 'timestamp', 'ht_result', 'points']
```

```python
ranking22=pd.concat([home_date22, away_date22]).sort_values(by=["home","timestamp"])
ranking22


ranking21=pd.concat([home_date21, away_date21]).sort_values(by=["home","timestamp"])
ranking21

ranking20=pd.concat([home_date20, away_date20]).sort_values(by=["home","timestamp"])
ranking20

ranking19=pd.concat([home_date19, away_date19]).sort_values(by=["home","timestamp"])
ranking19
```
