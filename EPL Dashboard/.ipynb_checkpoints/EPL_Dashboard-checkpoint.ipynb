```python
#IMPORT
import numpy as np
import pandas as pd
import bokeh
from bokeh.io import curdoc, show, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Grid, ImageURL, LinearAxis, Plot, Range1d, Select, Column, RangeTool, BoxSelectTool
from bokeh.models import ColumnDataSource, FactorRange, BoxSelectTool, CategoricalColorMapper,Tabs,TabPanel
from bokeh.layouts import column, row, grid
from bokeh.models import ColumnDataSource, FixedTicker
output_notebook()

```

```python
#this data is only for last season we want all 5 seasons

epl_teams=pd.read_csv("epl_teams.csv")
epl_only=epl_teams[epl_teams["tournament"]=="Premier League"]
epl_only['timestamp'] = pd.to_datetime(epl_only['date'])
epl_only22=epl_only[epl_only["timestamp"]>="2021-08-13"]
epl_only22=epl_only22[epl_only22["timestamp"]<="2022-05-22"]


teams=[i for i in epl_only["home"].unique()]


epl_only21=epl_only[epl_only["timestamp"]>="2020-09-12"]
epl_only21=epl_only21[epl_only21["timestamp"]<="2021-05-23"]


epl_only20=epl_only[epl_only["timestamp"]>="2019-08-9"]
epl_only20=epl_only20[epl_only20["timestamp"]<="2020-07-26"]
```

```python
home_date22=epl_only22[["home", "timestamp","ht_result"]]

home_date21=epl_only21[["home", "timestamp","ht_result"]]

home_date20=epl_only20[["home", "timestamp","ht_result"]]
```

```python
def points_ht(row):
    if row['ht_result'] == "WON":
        val = 3
    elif row['ht_result'] == "DRAW":
        val = 1
    else:
        val = 0
    return val
```

```python
def points_at(row):
    if row['at_result'] == "WON":
        val = 3
    elif row['at_result'] == "DRAW":
        val = 1
    else:
        val = 0
    return val
```

```python
#don't know error but it works
home_date22['points'] = home_date22.apply(points_ht, axis=1)
home_date21['points'] = home_date21.apply(points_ht, axis=1)
home_date20['points'] = home_date20.apply(points_ht, axis=1)
```

```python
home_date21
home_date22
home_date20
```

```python
away_date22=epl_only22[["away", "timestamp","at_result"]]
away_date22['points'] = away_date22.apply(points_at, axis=1)
away_date22.columns=['home', 'timestamp', 'ht_result', 'points']
away_date22


away_date21=epl_only21[["away", "timestamp","at_result"]]
away_date21['points'] = away_date21.apply(points_at, axis=1)
away_date21.columns=['home', 'timestamp', 'ht_result', 'points']
away_date21


away_date20=epl_only20[["away", "timestamp","at_result"]]
away_date20['points'] = away_date20.apply(points_at, axis=1)
away_date20.columns=['home', 'timestamp', 'ht_result', 'points']
```

```python
ranking22=pd.concat([home_date22, away_date22]).sort_values(by=["home","timestamp"])
ranking22


ranking21=pd.concat([home_date21, away_date21]).sort_values(by=["home","timestamp"])
ranking21

ranking20=pd.concat([home_date20, away_date20]).sort_values(by=["home","timestamp"])
ranking20
```

```python
standing22=ranking22.groupby(['home', 'timestamp']).sum("numeric_only") \
  .groupby(level=0).cumsum()

standing21=ranking21.groupby(['home', 'timestamp']).sum("numeric_only") \
  .groupby(level=0).cumsum()

standing20=ranking20.groupby(['home', 'timestamp']).sum("numeric_only") \
  .groupby(level=0).cumsum()
```

```python
ranking22['tot'] = ranking22.groupby(['home'])['points'].cumsum()
ranking22[ranking22["home"]=="Man City"]


ranking21['tot'] = ranking21.groupby(['home'])['points'].cumsum()
ranking21[ranking21["home"]=="Man City"]

ranking20['tot'] = ranking20.groupby(['home'])['points'].cumsum()
ranking20[ranking20["home"]=="Man City"]
```

```python
stadiums = pd.read_csv("stadiums-with-GPS-coordinates.csv")
stadiums=stadiums[stadiums["Country"]=="England"]
# stadiums
```

```python
from bokeh.tile_providers import Vendors
def wgs84_to_web_mercator_lon(lon):
    k = 6378137
    return lon * (k * np.pi/180.0)

def wgs84_to_web_mercator_lat(lat):
    k = 6378137
    return np.log(np.tan((90 + lat) * np.pi/360.0)) * k

stadiums["x"] = wgs84_to_web_mercator_lon(stadiums.Longitude)
stadiums["y"] = wgs84_to_web_mercator_lat(stadiums.Latitude)

source_stadium = ColumnDataSource(stadiums)


TOOLTIPS = [
    ("Team", "@Team"),
    ("City", "@City")
]

e = figure(x_range=(-448053.32822061086,145678.58558508987)
           , y_range=(6485356.016271952,7457123.475602542),
           height = 400, width = 550,            
           x_axis_type="mercator", y_axis_type="mercator",
          tooltips=TOOLTIPS)
e.xaxis.major_tick_line_color = None  # turn off x-axis major ticks
e.xaxis.minor_tick_line_color = None  # turn off x-axis minor ticks

e.yaxis.major_tick_line_color = None 
e.yaxis.minor_tick_line_color = None 
e.xaxis.major_label_text_color = None  # turn off x-axis tick labels leaving space
e.yaxis.major_label_text_color = None 

e.add_tile(Vendors.OSM)

e.circle(x='x', y='y', source=source_stadium, line_color="purple",
         size=10,fill_color='purple', alpha=0.5)

#show(e)
```
